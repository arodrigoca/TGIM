#Ej 1- Periodograma de AWGN


Calcule y represente el periodograma, utilizando Jupyter Notebook Tema 3,
de una señal x[n] de ruido AWGN de varianza unidad para N = 32. Tome Nfft =
1024. Cree para ello, en primer lugar una funci´on que compute el periodograma.
FFT es el número de muestras de la FFT. Una aproximación es poner el mismo número de muestras que las de la secuencia pricipal


```python
%matplotlib inline
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#periodogram definition
```

```python
def periodogram(x, nfft = 1024):
    """
    Computes periodogram of the sequence x using: 1/N|X(w)|²
    Where N is the number of samples of X
    """
    
    N = len(x)
    x_f  = np.fft.fft(x, nfft)
    Pxx = (np.abs(x_f)**2)/N
    
    Pxx = np.fft.fftshift(Pxx)
    
    f = np.fft.fftshift(np.fft.fftfreq(nfft))#el eje de frecuencias irá de -0.5 a 0.5
    
    return Pxx, f
```

```python
#generate AWGN
np.random.seed(1947)

N_sim = 10000
awgns = []
S_x = []

N = 32 #realization sample number

for i in range(N_sim):
    #generate awgn
    awgn = np.random.randn(N) #generate the sequence
    awgns.append(awgn)
    s_x, f = periodogram(awgn)
    S_x.append(s_x)
    
awgns = np.array(awgns)
print(awgns.shape)

S_x = np.array(S_x)
print(S_x.shape)

plt.figure(figsize=(8,6))
plt.plot(awgns.T, linestyle = ':', linewidth=0.5, color= [0.5, 0.5, 0.5])

m_awgns = np.mean(awgns, axis = 0)
std = np.std(awgns, axis = 0)

plt.plot(m_awgns, 'r', linewidth=2)
plt.plot(m_awgns+std, 'g', linewidth=2)
plt.plot(m_awgns-std, 'g', linewidth=2)


#sketch periodogram and mean periodogram

plot_2 = plt.figure()
plt.plot(f, 10*np.log10(S_x.T), linestyle = ':', linewidth=0.5, color= [0.5, 0.5, 0.5])
s_x_m = np.mean(S_x, axis=0)
plt.plot(f, 10*np.log10(s_x_m), 'r', linewidth=2) #es la esperanza del estimador S

s_x_v = np.var(S_x, axis=0)

plt.figure(figsize=(8,6))
plt.plot(f*2*np.pi, s_x_v)
```
